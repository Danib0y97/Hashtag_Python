Quais os produtos mais exportados ao longo de todo o período?

```python
produto_df = exp_franca[['Year' ,'SH4 Description', 'SH2 Description', "Net Weight"]] 

produto_df = produto_df.sort_values(by=['Net Weight'], ascending=False)



qnts = produto_df[["SH4 Description", 'SH2 Description']].value_counts()



#obs: Temos que utilizar no eixo X e Y o mesmo nome que está nas colunas!
#graf_bar = sns.barplot(data = produto_df, x = "SH4 Description" ,y= "Net Weight", color = "Blue")
#graf_bar.set_title("SH4")
#plt.show(graf_bar)

display(qnts)
#plt.show(graf_bar)
```

Em 2020, qual cidade mais exportou para a França?

```python


cidade_df = exp_franca.loc[exp_franca["Year"] == 2020]

cidade_df = cidade_df[['Year' ,'City', "Net Weight", "US$ FOB"]].groupby("City").sum()

cidade_df = cidade_df.sort_values(by=["US$ FOB"], ascending=False)
#df.loc[['viper', 'sidewinder'], ['shield']] = 50


display(cidade_df)
```

Quais os produtos mais exportados que as 2 maiores cidade exportam?

```python
cidade = cidade_df.index[0]

cidade_df = exp_franca.loc[exp_franca["City"] == cidade]

prod_cidade = cidade_df[['SH2 Description', "US$ FOB"]].groupby("SH2 Description").sum()
prod_cidade = prod_cidade.sort_values(by=["US$ FOB"], ascending=False)

#prod_cidade = 


display(prod_cidade)
```
